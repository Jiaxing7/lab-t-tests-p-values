### 1. One tailed t-test 

```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ttest_machine.txt', sep = ' ')
df
```

```python
st.ttest_rel(df['New_machine'], df['Old_machine'])
```

```python
# p-value < 0.05  new machine's mean time !=43.23
# statistic is negative.  It means the times New machine < 43.23(mean time from Old machine). 
# Provide sufficient evidence to show new machine is better
```

### 2.Matched Pairs Test

```python
df1 = pd.read_csv('pokemon.txt', sep=',')
df1
```

```python
st.ttest_ind(df1['Attack'],df1['Defense'])
```

```python
# p-value < 0.05, reject H0, there are significant differences of Attack vs Defense
```
